* Integração numérica: Quadratura Gauss-Legendre
* Desafio: acurácia
* Número de pontos quadratura ~ acurácia
* Distância ponto observação ~ acurácia (Ku, 1977)
    * d < distância entre pontos
* Discretização adaptativa (Li et al., 2011)

# Discretização adaptativa

```python
Image(filename='images/tesseroid-distance-size.png', width=400)
```

1. Fixar N x N x N pontos (eu uso N=2):
2. L = maior dimensão. d < L ?
3. Divide em 8
4. Repete para sub-tesseroides enquanto violar 2.

```python
def adaptive_discretize(tesseroid, point, ratio):
    lon, lat, h = point
    d2r = np.pi/180.    
    radius = h + MEAN_EARTH_RADIUS
    mesh = []
    queue = [tesseroid]
    while queue:
        tesseroid = queue.pop()
        tes_radius = tesseroid.top + MEAN_EARTH_RADIUS
        tes_lat = d2r*0.5*(tesseroid.s + tesseroid.n)
        tes_lon = d2r*0.5*(tesseroid.w + tesseroid.e)
        distance = np.sqrt(radius**2 + tes_radius**2 - 2.*radius*tes_radius*(
            np.sin(d2r*lat)*np.sin(tes_lat) + np.cos(d2r*lat)*np.cos(tes_lat)*np.cos(d2r*lon - tes_lon)))
        sizes = [MEAN_EARTH_RADIUS*d2r*(tesseroid.e - tesseroid.w),
                 MEAN_EARTH_RADIUS*d2r*(tesseroid.n - tesseroid.s),
                 tesseroid.top - tesseroid.bottom]
        splits = [2 if distance < ratio*size else 1 for size in sizes]
        if 2 in splits:
            queue.extend(tesseroid.half())
        else:
            mesh.append(tesseroid)
    return mesh

def check(tesseroids, point, ratio):
    lon, lat, h = point
    d2r = np.pi/180.    
    radius = h + MEAN_EARTH_RADIUS
    for tesseroid in tesseroids:
        tes_radius = tesseroid.top + MEAN_EARTH_RADIUS
        tes_lat = d2r*0.5*(tesseroid.s + tesseroid.n)
        tes_lon = d2r*0.5*(tesseroid.w + tesseroid.e)
        distance = np.sqrt(radius**2 + tes_radius**2 - 2.*radius*tes_radius*(
            np.sin(d2r*lat)*np.sin(tes_lat) + np.cos(d2r*lat)*np.cos(tes_lat)*np.cos(d2r*lon - tes_lon)))        
        sizes = [MEAN_EARTH_RADIUS*d2r*(tesseroid.e - tesseroid.w),
                 MEAN_EARTH_RADIUS*d2r*(tesseroid.n - tesseroid.s),
                 tesseroid.top - tesseroid.bottom]
        if False in [distance >= ratio*size for size in sizes]:
            return False        
    return True
```

# Quantificar erro

* Comparar solução analítica
    * Meia casca esférica 
    * P no polo
    * Discretizar em tesseroides
* d < L => d < R*L
* Erro X R
* Determinar R para erro = 1%

```python
def make_model(size, thickness=50000, density=2700):
    # Stream the model to avoid overloading the memory
    for w in np.arange(0, 180, size):
        for s in np.arange(-90, 90, size):
            yield Tesseroid(w, w+size, s, s+size, 0, -thickness, {'density':density})
```

# Resultados

```python
shelleffect = {'pot':gravmag.half_sph_shell.potential, 
               'gz':gravmag.half_sph_shell.gz, 
               'gxx':gravmag.half_sph_shell.gxx, 
               'gyy':gravmag.half_sph_shell.gyy, 
               'gzz':gravmag.half_sph_shell.gzz}
sizes = [0.1, 1, 4, 10, 20] # in degrees
ratios = range(9)
heights = array([1, 100, 1000, 10000, 100000, 250000, 500000]) # in meters
lons = 90*ones_like(heights)
lats = zeros_like(heights)
thickness = 50000
density = 2700
fields = ['pot', 'gz', 'gxx', 'gxy', 'gxz', 'gyy', 'gyz', 'gzz']
```

```python
result_comparison = dict((f, [{'size':s, 'ratio/height':[]} for s in sizes]) for f in fields)
for i, size in enumerate(sizes):    
    print 'Tesseroids of size %g' % (size)
    for ratio in ratios:
        # Remove comments and convert the strings to floats
        output = [gravmag.tesseroid.potential(lons, lats, heights, make_model(size), ratio=ratio),
                  gravmag.tesseroid.gz(lons, lats, heights, make_model(size), ratio=ratio),
                  gravmag.tesseroid.gxx(lons, lats, heights, make_model(size), ratio=ratio),
                  gravmag.tesseroid.gxy(lons, lats, heights, make_model(size), ratio=ratio),
                  gravmag.tesseroid.gxz(lons, lats, heights, make_model(size), ratio=ratio),
                  gravmag.tesseroid.gyy(lons, lats, heights, make_model(size), ratio=ratio),
                  gravmag.tesseroid.gyz(lons, lats, heights, make_model(size), ratio=ratio),
                  gravmag.tesseroid.gzz(lons, lats, heights, make_model(size), ratio=ratio)]
        # Place the output in the result dict
        for f, field in enumerate(fields):
            result_comparison[field][i]['ratio/height'].append(output[f])
```

```python
for field in fields:
    for i in xrange(len(sizes)):
        for r in xrange(len(ratios)):
            result_comparison[field][i]['ratio/height'][r] = result_comparison[field][i]['ratio/height'][r].tolist()
with open('result_comparison.json', 'w') as f:
    json.dump(result_comparison, f, indent=1)
```

```python
def make_plots(results, save=True):
    titles = ['$Potential$', r'$g_{z}$', r'$g_{xx}$', r'$g_{yy}$', r'$g_{zz}$', r'$g_{xy}$', r'$g_{xz}$', r'$g_{yz}$']
    fields = ['pot', 'gz', 'gxx', 'gyy', 'gzz', 'gxy', 'gxz', 'gyz']
    for t, field in zip(titles, fields):
        figure(figsize=(6, 5))
        suptitle(t, fontsize=16)
        subplots_adjust(bottom=0.1, hspace=0.29, top=0.88, wspace=0.33, left=0.12, right=0.98)
        for i in xrange(len(sizes)):
            ax = subplot(2, 3, i + 1)
            title(r'Size: %g$^\circ$' % (sizes[i]), fontsize=12)
            for j, height in enumerate(heights):
                if field in ['gxy', 'gxz', 'gyz']:
                    difference = abs(transpose(results[field][i]['ratio/height'])[j])
                else:
                    shell = shelleffect[field](height, 0, -thickness, density)
                    difference = 100*abs((transpose(results[field][i]['ratio/height'])[j] - shell)/float(shell))
                plot(ratios, difference, '-', label='%g' % (0.001*height))
            hlines([1], min(ratios), max(ratios), 'r', '--')
            ax.set_yscale('log')           
            xlim(min(ratios), max(ratios))
            ylim(10**(-11), 10**4)
            if field in ['gxy', 'gxz', 'gyz']:
                ylim(10**(-19), 10)
                if i in [0, 3]:
                    ylabel(r'Error ($E\"otv\"os$)')
            else:
                if i in [0, 3]:
                    ylabel('Error (%)')
            if i in [3, 4, 5]:
                xlabel('Distance/size')
            ax.set_yticks(ax.get_yticks()[1:])
            grid()
        legend(title='Height (km)', shadow=True, fancybox=True, ncol=1, prop={'size':12}, loc=(1.4, -0.2))
        if save:
            savefig('images/diff-tess-shell-%s.pdf' % (field))
            savefig('images/diff-tess-shell-%s.png' % (field))
```
