```python
%pylab inline
import numpy as np
from IPython.display import Image
from fatiando import gravmag, utils, gridder
from fatiando.mesher import Tesseroid, TesseroidMesh, vremove
from fatiando.vis import mpl, myv
```

# Cadeia meso-oceânica

```python
def snapshots(scene, fname):
    scene.scene.camera.position = [39328027.69104398, -9396322.0834418982, -19867034.915217586]
    scene.scene.camera.focal_point = [2048089.5665244255, -940170.49195667217, -974542.32651143684]
    scene.scene.camera.view_angle = 15.35999999999999
    scene.scene.camera.view_up = [0.39034841733481618, -0.25534037990498637, 0.88455039623199616]
    scene.scene.camera.clipping_range = [25373516.662868574, 69926182.012640104]
    scene.scene.camera.compute_view_plane_normal()
    scene.scene.render()
    myv.savefig('images/%s.png' % (fname))
    scene.scene.camera.position = [-15153055.631171998, 8338256.0942435581, -36705585.026986741]
    scene.scene.camera.focal_point = [6472981.216103361, -1403080.1936048171, -1270468.8266999177]
    scene.scene.camera.view_angle = 0.5404319552844592
    scene.scene.camera.view_up = [0.83246763208743546, -0.11975739897630805, -0.54097671568854744]
    scene.scene.camera.clipping_range = [19818060.668792952, 65711713.470129766]
    scene.scene.camera.compute_view_plane_normal()
    scene.scene.render()
    myv.savefig('images/%s2.png' % (fname))
```

```python
s, n, top, bottom = -50, 0, -1000, -15000
middles = np.linspace(-20, -5, 50)[::-1]
dlons = np.linspace(0.1, 2, 8)
souths = np.linspace(s, n, 50)
tops = np.linspace(bottom, top, len(dlons))[::-1]
model = [Tesseroid(middles[j] - 0.5*dlons[i], middles[j] + 0.5*dlons[i], souths[j], souths[j+1], tops[i], tops[i+1], {'density':200})
         for i in range(len(dlons)-1) for j in range(len(souths)-1)]
print abs(tops[1] - tops[0])
```

```python
scene = myv.figure(zdown=False, size=(800,600))
myv.tesseroids(model, 'density', edges=True)
myv.earth(opacity=0.4)
myv.core(opacity=1)
myv.continents(linewidth=1.5)
snapshots(scene, 'ridge-model')
myv.show()
```

```python
bounds = [-30, 10, -60, 10, -1000, -20000]
shape = (100, 100)
lons, lats, heights = gridder.regular(bounds[:4], shape, z=250000)
gz = utils.contaminate(gravmag.tesseroid.gz(lons, lats, heights, model), 0.1)

mpl.figure(figsize=(12, 10))
bm = mpl.basemap(bounds[:4], 'ortho')
bm.drawcoastlines()
bm.drawmapboundary()
mpl.contourf(lons, lats, gz, shape, 50, basemap=bm)
cb = mpl.colorbar()
cb.set_label('mGal')
mpl.savefig('images/ridge-gravity.png', dpi=500)
```

```python
mesh = TesseroidMesh(bounds, (20, 100, 300))
slons = middles[::2]
slats = np.linspace(-49.5, -0.5, len(slons))
sdepths = -5000*np.ones_like(slons)
props = model[0].props
seeds = gravmag.harvester.sow([[x, y, z, props] for x, y, z in zip(slons, slats, sdepths)], mesh)
```

```python
scene = myv.figure(zdown=False, size=(800, 600))
myv.tesseroids(model, style='wireframe', linewidth=2, color=(0,0,0))
myv.tesseroids(seeds)
myv.earth(opacity=0.4)
myv.core(opacity=1)
myv.continents(linewidth=1.5)
snapshots(scene, 'ridge-seed')
myv.show()
```

```python
data = [gravmag.harvester.Gz(lons, lats, heights, gz, meshtype='tesseroid')]
estimate, predicted = gravmag.harvester.harvest(data, seeds, mesh, compactness=10, threshold=0.0001)
mesh.addprop('density', estimate['density'])
ridge = vremove(0, 'density', mesh)
print len(ridge)
```

```python
mpl.figure(figsize=(12,10))
bm = mpl.basemap(bounds[:4], 'ortho')
bm.drawcoastlines()
bm.drawmapboundary()
levels = mpl.contour(lons, lats, gz, shape, 10, basemap=bm)
mpl.contour(lons, lats, predicted[0], shape, levels, basemap=bm, color='r')
mpl.savefig('images/ridge-predicted.png', dpi=500)
```

```python
scene = myv.figure(zdown=False, size=(800, 600))
myv.tesseroids(model, style='wireframe', linewidth=2, color=(0,0,0))
myv.tesseroids(ridge)
myv.earth(opacity=0.4)
myv.core(opacity=1)
myv.continents(linewidth=1.5)
snapshots(scene, 'ridge-estimate')
myv.show()
```
