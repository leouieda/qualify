```python
%pylab inline
import numpy as np
from IPython.display import Image
from fatiando import gravmag, utils, gridder
from fatiando.mesher import Tesseroid, TesseroidMesh, vremove
from fatiando.vis import mpl, myv
```

## Placa em subducção

```python
def subduction_snapshots(scene, fname):
    scene.scene.camera.position = [14739971.720808242, -34793228.438286372, -27937201.218419865]
    scene.scene.camera.focal_point = [1661310.7297087216, -3566782.5101824086, -2012613.2004857999]
    scene.scene.camera.view_angle = 9.8304
    scene.scene.camera.view_up = [0.35323133837607151, -0.50556886970345638, 0.78716436630212061]
    scene.scene.camera.clipping_range = [25636419.335725315, 73973736.575326011]
    scene.scene.camera.compute_view_plane_normal()
    scene.scene.render()
    myv.savefig('images/%s.png' % (fname))
    scene.scene.camera.position = [1279596.3804397113, 14129333.406946078, -40822540.77722314]
    scene.scene.camera.focal_point = [1768487.1549586861, -4787883.3357389905, -2610939.671781708]
    scene.scene.camera.view_angle = 3.2212254719999995
    scene.scene.camera.view_up = [0.39231718616903671, -0.8223362063030808, -0.41212909293019595]
    scene.scene.camera.clipping_range = [25271154.906635709, 65049900.94522959]
    scene.scene.camera.compute_view_plane_normal()
    scene.scene.render()
    myv.savefig('images/%s2.png' % (fname))
```

```python
w, e, s, n, top, bottom = -75, -68, -50, -20, -30000, -300000
wests = np.linspace(w, e, 8)
souths = np.linspace(s, n, 20)
dlon = wests[1] - wests[0]
bottoms = np.linspace(bottom, top, len(wests))[::-1]
subduction_model = [Tesseroid(wests[i] - 0.7*dlon, wests[i+1], souths[j], souths[j+1], bottoms[i], bottoms[i+1], {'density':-200})
                    for i in range(len(wests)-1) for j in range(len(souths)-1)]
print abs(bottoms[1] - bottoms[0])
```

```python
scene = myv.figure(zdown=False, size=(800,600))
myv.tesseroids(subduction_model, 'density')
myv.earth(opacity=0.4)
myv.core(opacity=1)
myv.continents(linewidth=1.5)
subduction_snapshots(scene, 'subduction-model')
myv.show()
```

```python
bounds = [-100, -40, -60, -10, -30000, -800000]
shape = (100, 100)
lons, lats, heights = gridder.regular(bounds[:4], shape, z=250000)
gz = utils.contaminate(gravmag.tesseroid.gz(lons, lats, heights, subduction_model), 0.1)

mpl.figure(figsize=(12, 10))
bm = mpl.basemap(bounds[:4], 'ortho')
bm.drawcoastlines()
bm.drawmapboundary()
mpl.contourf(lons, lats, gz, shape, 50, basemap=bm)
cb = mpl.colorbar()
cb.set_label('mGal')
mpl.savefig('images/subduction-gravity.png', dpi=500)
```

```python
mesh = TesseroidMesh(bounds, (20, 20, 60))
props = subduction_model[0].props

seeds = gravmag.harvester.sow([[-74, -35, -30000, props],
                               [-71, -35, -300000, props],
                               [-69, -35, -500000, props],
                               [-74, -25, -30000, props],
                               [-71, -25, -300000, props],
                               [-69, -25, -500000, props],
                               [-74, -45, -30000, props],
                               [-71, -45, -300000, props],
                               [-69, -45, -500000, props]], mesh)
seeds = gravmag.harvester.sow([[-74, -35, -30000, props]], mesh)
seeds = gravmag.harvester.sow([[-74, -45, -30000, props],
                               [-74, -40, -30000, props],
                               [-74, -35, -30000, props],
                               [-74, -30, -30000, props],
                               [-74, -25, -30000, props]], mesh)
```

```python
scene = myv.figure(zdown=False, size=(800, 600))
myv.tesseroids(subduction_model, style='wireframe', linewidth=2, color=(0,0,0))
myv.tesseroids(seeds)
myv.earth(opacity=0.4)
myv.core(opacity=1)
myv.continents(linewidth=1.5)
subduction_snapshots(scene, 'subduction-seed')
myv.show()
```

```python
data = [gravmag.harvester.Gz(lons, lats, heights, gz, meshtype='tesseroid')]
estimate, predicted = gravmag.harvester.harvest(data, seeds, mesh, compactness=10, threshold=0.00001)
mesh.addprop('density', estimate['density'])
subduction_estimate = vremove(0, 'density', mesh)
print len(subduction_estimate)
```

```python
mpl.figure(figsize=(12,10))
bm = mpl.basemap(bounds[:4], 'ortho')
bm.drawcoastlines()
bm.drawmapboundary()
levels = mpl.contour(lons, lats, gz, shape, 10, basemap=bm)
mpl.contour(lons, lats, predicted[0], shape, levels, basemap=bm, color='r')
mpl.savefig('images/subdution-predicted.png', dpi=500)
```

```python
scene = myv.figure(zdown=False, size=(800, 600))
myv.tesseroids(subduction_model, style='wireframe', linewidth=2, color=(0,0,0))
myv.tesseroids(subduction_estimate)
myv.earth(opacity=0.4)
myv.core(opacity=1)
myv.continents(linewidth=1.5)
subduction_snapshots(scene, 'subduction-estimate')
myv.show()
```
