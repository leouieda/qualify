```python
def snapshots(scene, fname):
    scene.scene.camera.position = [31037168.278975949, -33856809.457610652, -19419950.493962821]
    scene.scene.camera.focal_point = [4536775.6812402923, -4870469.8950687684, -2814292.6192687405]
    scene.scene.camera.view_angle = 9.830399999999985
    scene.scene.camera.view_up = [0.19004298158576247, -0.35142721651001563, 0.91672382790348417]
    scene.scene.camera.clipping_range = [27934714.278328925, 77595563.48563087]
    scene.scene.camera.compute_view_plane_normal()
    scene.scene.render()
    myv.savefig('images/%s.png' % (fname))
    scene.scene.camera.position = [-14264387.499854848, 8922615.8404255956, -38807129.189956002]
    scene.scene.camera.focal_point = [3643755.5910187904, -4457862.7933716653, -2496142.3528043735]
    scene.scene.camera.view_angle = 0.6755399441055734
    scene.scene.camera.view_up = [0.51794833496497172, -0.68761923346765808, -0.50883131986186114]
    scene.scene.camera.clipping_range = [21309576.046163626, 67777981.071953371]
    scene.scene.camera.compute_view_plane_normal()
    scene.scene.render()
    myv.savefig('images/%s2.png' % (fname))
```

```python
model = [Tesseroid(-53, -48, -22.5, -17.5, -30*10**3, -38*10**3, {'density':200}),
         Tesseroid(-52, -49, -21.5,  -18.5, -38*10**3, -45*10**3, {'density':200})]
```

```python
scene = myv.figure(zdown=False, size=(800,600))
myv.tesseroids(model, 'density', edges=True)
myv.earth(opacity=0.4)
myv.core(opacity=1)
myv.continents(linewidth=1.5)
snapshots(scene, 'underplating-model')
myv.show()
```

```python
bounds = [-57, -44, -26, -14, -30*10**3, -50*10**3]
shape = (100, 100)
lons, lats, heights = gridder.regular(bounds[:4], shape, z=100000)
gz = utils.contaminate(gravmag.tesseroid.gz(lons, lats, heights, model), 0.1)
```

```python
mpl.figure()
bm = mpl.basemap(bounds[:4], 'merc')
bm.drawcoastlines()
bm.drawmapboundary()
mpl.contourf(lons, lats, gz, shape, 50, basemap=bm)
cb = mpl.colorbar()
cb.set_label('mGal')
mpl.savefig('images/underplating-gravity.png', dpi=500)
```

```python
mesh = TesseroidMesh(bounds, (20, 30, 30))
props = model[0].props
seeds = gravmag.harvester.sow([[-50.5, -20.5, -30*10**3, props]], mesh)
print mesh.dims
```

```python
scene = myv.figure(zdown=False, size=(800, 600))
myv.tesseroids(model, style='wireframe', linewidth=2, color=(0,0,0), scale=(1, 1, 10))
myv.tesseroids(seeds, scale=(1, 1, 10))
snapshot_exaggerate(scene, 'underplating-seed')
myv.show()
```

```python
data = [gravmag.harvester.Gz(lons, lats, heights, gz, meshtype='tesseroid')]
estimate, predicted = gravmag.harvester.harvest(data, seeds, mesh, compactness=1, threshold=0.0001)
mesh.addprop('density', estimate['density'])
underplate = vremove(0, 'density', mesh)
print len(underplate)
```

```python
mpl.figure()
bm = mpl.basemap(bounds[:4], 'merc')
bm.drawcoastlines()
bm.drawmapboundary()
levels = mpl.contour(lons, lats, gz, shape, 10, basemap=bm)
mpl.contour(lons, lats, predicted[0], shape, levels, basemap=bm, color='r')
mpl.savefig('images/underplating-predicted.png', dpi=500)
```

```python
scene = myv.figure(zdown=False, size=(800, 600))
myv.tesseroids(model, style='wireframe', linewidth=2, color=(0,0,0), scale=(1, 1, 10))
myv.tesseroids(underplate, scale=(1, 1, 10))
snapshot_exaggerate(scene, 'underplating-estimate')
myv.show()
```
