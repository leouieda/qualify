## "Pluma" mantélica

### O modelo

```python
def plume_snapshots(scene, fname):
    scene.scene.camera.position = [34857053.692156821, -16244344.702579584, -19621678.559786443]
    scene.scene.camera.focal_point = [217919.00569067523, -2531579.6254882161, 1122404.5636951104]
    scene.scene.camera.view_angle = 7.864320000000001
    scene.scene.camera.view_up = [0.38901581666244622, -0.32265376770943532, 0.86287962113457961]
    scene.scene.camera.clipping_range = [22103149.499328904, 69602532.987363279]
    scene.scene.camera.compute_view_plane_normal()
    scene.scene.render()
    myv.savefig('images/%s.png' % (fname))
    scene.scene.camera.position = [32029777.544032086, 25237005.526905328, -119433.5809351519]
    scene.scene.camera.focal_point = [1828521.8650542684, -4863182.7349342797, -418406.04682270699]
    scene.scene.camera.view_angle = 5.033164800000001
    scene.scene.camera.view_up = [0.70366336903717563, -0.70515972339999078, -0.087221715012377854]
    scene.scene.camera.clipping_range = [22083364.018758003, 63780192.676591538]
    scene.scene.camera.compute_view_plane_normal()
    scene.scene.render()
    myv.savefig('images/%s2.png' % (fname))
```

```python
plume_model = [Tesseroid(-50, -40, -10, -0, -100*10**3, -250*10**3, {'density':-100}),
               Tesseroid(-48, -42, -8, -2, -250*10**3, -400*10**3, {'density':-100}),
               Tesseroid(-46, -44, -6, -4, -300*10**3, -2700*10**3, {'density':-100})]
```

```python
scene = myv.figure(zdown=False, size=(800,600))
myv.tesseroids(plume_model, 'density')
myv.earth(opacity=0.4)
myv.core(opacity=1)
myv.continents(linewidth=1.5)
plume_snapshots(scene, 'plume-model')
myv.show()
```

### Dados

```python
bounds = [-60, -30, -20, 10, 0, -3000*10**3]
shape = (100, 100)
lons, lats, heights = gridder.regular(bounds[:4], shape, z=10000)
gz = utils.contaminate(gravmag.tesseroid.gz(lons, lats, heights, plume_model), 0.01, percent=True)

mpl.figure()
bm = mpl.basemap(bounds[:4], 'merc')
bm.drawcoastlines()
bm.drawmapboundary()
mpl.contourf(lons, lats, gz, shape, 50, basemap=bm)
cb = mpl.colorbar()
cb.set_label('mGal')
mpl.savefig('images/plume-gravity.png', dpi=500)
```

### Escolha da semente

```python
mesh = TesseroidMesh(bounds, (60, 60, 60))
seeds = gravmag.harvester.sow([[-45, -5, -100000, plume_model[0].props]], mesh)
print mesh.dims
```

```python
scene = myv.figure(zdown=False, size=(800, 600))
myv.tesseroids(plume_model, style='wireframe', linewidth=2)
myv.tesseroids(seeds, color=(0, 1, 0))
myv.earth(opacity=0.4)
myv.core(opacity=1)
myv.continents(linewidth=1.5)
plume_snapshots(scene, 'plume-seed')
myv.show()
```

### Inversão

```python
data = [gravmag.harvester.Gz(lons, lats, heights, gz, meshtype='tesseroid')]
estimate, predicted = gravmag.harvester.harvest(data, seeds, mesh, compactness=10, threshold=0.000005)
mesh.addprop('density', estimate['density'])
plume_estimate = vremove(0, 'density', mesh)
print len(plume_estimate)
```

```python
mpl.figure()
bm = mpl.basemap(bounds[:4], 'merc')
bm.drawcoastlines()
bm.drawmapboundary()
levels = mpl.contour(lons, lats, gz, shape, 10, basemap=bm)
mpl.contour(lons, lats, predicted[0], shape, levels, basemap=bm, color='r')
mpl.savefig('images/plume-predicted.png', dpi=500)
```

```python
scene = myv.figure(zdown=False, size=(800, 600))
myv.tesseroids(plume_model, style='wireframe', linewidth=3, color=(0,0,0))
myv.tesseroids(plume_estimate)
myv.earth(opacity=0.4)
myv.core(opacity=1)
myv.continents(linewidth=1.5)
plume_snapshots(scene, 'plume-estimate')
myv.show()
```
